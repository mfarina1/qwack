```python
class Ball(pygame.sprite.Sprite):
    def __init__(self, leftCircleX, leftCircleY, ID, color, width):
        super().__init__()
        self.circle = pygame.Rect(leftCircleX,leftCircleY,width,width)
        self.ID = ID
        self.color = color
        self.observedCount = 1
        # Velocity
        self.change_x = random.randint(1,3)
        self.change_y = random.randint(1,3)
        self.panelSide = ""
        #print(self.circle.midtop)
#     def checkCollision(self, panel):
#         if self.circle.colliderect(panel):
#             #Check which side
            
    def update(self):
        self.circle.y += self.change_y
        self.circle.x += self.change_x
        self.panelSide = ""
        
        if self.observedCount == 0:
            self.change_x = random.randint(-3,3)
            self.change_y = random.randint(-3,3)
        self.observedCount += 1
```

```python
def randomValue():
    return random.randint(0, 255)
```

```python
def drawCircle(screen, COLOR, x, y, size):
    
    rect = pygame.draw.circle(screen, COLOR, (x, y), size)
    return rect
```

```python
def checkCollision(self, panel):
#     print(self.circle.midtop[1]-panel.midbottom[1])
    if self.circle.colliderect(panel):
        if abs(self.circle.midtop[1]-panel.midbottom[1]) < 30:
            self.panelSide = "BOTTOM"
        elif abs(self.circle.midright[0]-panel.midleft[0]) < 30:
            self.panelSide = "LEFT"
        elif abs(self.circle.midleft[0]-panel.midright[0]) < 30:
            self.panelSide = "RIGHT"
        elif abs(self.circle.midbottom[1]-panel.midtop[1]) < 30:
            self.panelSide = "TOP"
```
