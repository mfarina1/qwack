```python
def randomValue():
    return random.randint(0, 255)
```

```python
def drawCircle(screen, COLOR, x, y, size):
    
    rect = pygame.draw.circle(screen, COLOR, (x, y), size)
    return rect
```

```python
def checkCollision(self, panel):
#     print(self.circle.midtop[1]-panel.midbottom[1])
    if self.circle.colliderect(panel):
        if abs(self.circle.midtop[1]-panel.midbottom[1]) < 30:
            self.panelSide = "BOTTOM"
        elif abs(self.circle.midright[0]-panel.midleft[0]) < 30:
            self.panelSide = "LEFT"
        elif abs(self.circle.midleft[0]-panel.midright[0]) < 30:
            self.panelSide = "RIGHT"
        elif abs(self.circle.midbottom[1]-panel.midtop[1]) < 30:
            self.panelSide = "TOP"
```
