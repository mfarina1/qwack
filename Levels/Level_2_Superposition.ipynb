```python
def level2():
    duckHitbox = pygame.Rect(0,0,64, 64)
    xChange = 0
    yChange = 0
    y = 450
    x = 100
    running = True
    while running:
        screen.fill((0,0,0))
        
        mx, my = pygame.mouse.get_pos()
        
        platform = pygame.Rect(0,400,800,200)
        pygame.draw.rect(screen, (112,128,144), platform)
        
        

        click = False
        for event in pygame.event.get():
            if event.type == QUIT:
                pygame.quit()
                sys.exit()
            if event.type == KEYDOWN:
                if event.key == K_ESCAPE:
                    running = False
                if event.key == pygame.K_LEFT:
                    xChange = -5
                if event.key == pygame.K_RIGHT:
                    xChange = 5
                if event.key == pygame.K_UP:
                    yChange = -5
                if event.key == pygame.K_DOWN:
                    yChange = 5
            if event.type == MOUSEBUTTONDOWN:
                if event.button == 1:
                    click = True 
                    print(mx, my)
            if event.type == pygame.KEYUP:
                if event.key == pygame.K_LEFT or event.key == pygame.K_RIGHT:
                    xChange = 0
                if event.key == pygame.K_UP or event.key == pygame.K_DOWN:
                    yChange = 0
                    
                    
        x += xChange
        if x <= 0:
            x = 0
        elif x >= 800-duckHitbox.width:
            x = 800-duckHitbox.width

        y += yChange
        if y <= 400:
            y = 400
        elif y >= 600-duckHitbox.height:
            y = 600-duckHitbox.height
        duckHitbox = Icon(icon, x, y)
        pygame.display.update()
        mainClock.tick(120) 
```
